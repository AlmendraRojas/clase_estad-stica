### Considerar mediana y media , elegir una de ellas como la probabilidad de éxito

### Primera pregunta, tenemos simetria?

```r
hist(efi)
```

### Vemos en el histograma que no hay simetria, 
### por tanto las medidas de tendencia central (media, mediana y moda) SON diferentes.

```r
# Calcuamos el promedio (media)
miMedia <- mean(efi)
```

```r
### Nos da NaN (not a number); esto es por los valores perdidos....
miMedia
```

## Media

```r
# Calcuamos el promedio (media) y le especificamos que DESCARTE los valores perdidos
# usamos la opcion na.rm=TRUE
miMedia <- mean(efi, na.rm=TRUE)
miMedia
```

## Mediana

```r
# Calculamos la mediana igual...
miMediana <- median(efi, na.rm=TRUE)
miMediana
```

---
#### Nuestros datos tienen un sesgo por la asimetría (ver gráfico de histograma); por esta razón
#### no es recomendable usar la media.
#### No es una generalidad ni una "regla" usar la mediana, simplemente se recomienda cuando tenemos un sesgo como el observado en estos datos
---

##### Volviendo a nuestro ejercicio...

## Usaremos la mediana
#### Probabilidad de éxito:

```r
pExito <- miMediana
pExito
```

#### Probabilidad de fracazo:

```r
pFracazo <- 1- miMediana
pFracazo
```

### Redondeados


```r
round(miMediana,4)
```

```r
round(1- miMediana, 4)
```

### Primera vuelta

```r
exito1 <-round(pExito,4)
fracazo1 <-round(pFracazo,4)

cat(c("En la primera vuelta, la probabilidad de exito es", exito1, "y la de fracazo es ",fracazo1))
```

### Segunda vuelta

```r
fracazo2 <- round(fracazo1 * fracazo1,4)
exito2 <- 1 - fracazo2

cat(c("En la segunda vuelta, la probabilidad de exito es", exito2, "y la de fracazo es ",fracazo2))
```

### Tercera vuelta

```r
fracazo3 <- round(fracazo2 * fracazo1,4)
exito3 <- 1 - fracazo3

cat(c("En la segunda vuelta, la probabilidad de exito es", exito3, "y la de fracazo es ",fracazo3))
```
